```python
author_list = []
narrator_list = []
audio_lengths = []
overall_ratings = []
performance_ratings = []
story_ratings = []
num_ratings_list = []
skipped_links_indices = []

for index, link in enumerate(audible_links_data['url']):

    r = requests.get('https://www.audible.in/' + link)
    soup = BeautifulSoup(r.content, 'html5lib')

    metadata = soup.find_all("div", {"id": "center-1"})
    
    try:
        for i in metadata:
            metadata_container = i.find_all('div', {'class': "bc-container"})

        for i in metadata_container:
            if i== None:
                span = i.find('div', {'class': "bc-col-responsive bc-col-5"}).find('span')

        author_list.append(span.find('li', {"class": "authorLabel"}).a.text)
        narrator_list.append(span.find('li', {"class": "narratorLabel"}).a.text)
        audio_lengths.append(((span.find('li', {"class": "runtimeLabel"}).text).replace('\n', '')).split(':')[-1].strip())
        num_ratings_list.append(span.find('li', {'class': 'ratingsLabel'}).text.replace('\n', '').replace('\t', '').split('(')[-1].split(' ')[0])

        reviews_section = soup.find_all("div", {"id": "center-16"})
        for i in reviews_section:
            metadata_container = i.find_all('div', {'class': "bc-container"})

        for i in metadata_container:
            star_section = i.find_all('div', {'class': 'bc-row-responsive bc-spacing-s6'})

        for i in star_section:
            overall_perf_story = i.find_all('div', {'class':'bc-col-responsive bc-col-4'})

        overall_section = overall_perf_story[0].find('span').find_all('li', {'class': 'rating-stars'})
        for i in overall_section:
            overall_ratings.append(i.find_all('span')[-1].text.strip().split(' ')[0])

        perf_section = overall_perf_story[1].find('span').find_all('li', {'class': 'rating-stars'})
        for i in perf_section:
            performance_ratings.append(i.find_all('span')[-1].text.strip().split(' ')[0])

        story_section = overall_perf_story[2].find('span').find_all('li', {'class': 'rating-stars'})
        for i in story_section:
            story_ratings.append(i.find_all('span')[-1].text.strip().split(' ')[0])
    except Exception as e:
        print('Skipping', e)
        skipped_links_indices.append(index)
        continue
```
